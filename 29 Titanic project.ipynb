```python
def age_filter(str):
    if str == "male":
        return 1
    else:
        return 2

titanic["Gender"] = titanic["Sex"].apply(age_filter)
titanic.head()
```

```python
mean_1 = titanic[titanic.Survived == 1].Age.mean()
mean_1
```

```python
titanic["age"] = np.where(pd.isnull(titanic.Age) & titanic.Survived == 1,mean_1,titanic.Age)
titanic.head()
```

```python
mean_0 = titanic[titanic.Survived==0].Age.mean()
mean_0
```

```python
male = (titanic.Sex ==1).sum()
female = (titanic.Sex ==2).sum()
plt.pie([male,female],labels=["Male","Female"],colors=["red","yellow"],explode=(0.10,0),startangle=90)
plt.axis("equal")
plt.show()
```

```python

MaleS=titanic[titanic.Sex==1][titanic.Survived==1].shape[0]
print(MaleS)
MaleN=titanic[titanic.Sex==1][titanic.Survived==0].shape[0]
print(MaleN)
FemaleS=titanic[titanic.Sex==2][titanic.Survived==1].shape[0]
print(FemaleS)
FemaleN=titanic[titanic.Sex==2][titanic.Survived==0].shape[0]
print(FemaleN)
chart=[MaleS,MaleN,FemaleS,FemaleN]
colors=['lightskyblue','yellowgreen','Yellow','Orange']
labels=["Survived Male","Not Survived Male","Survived Female","Not Survived Female"]
explode=[0,0.05,0,0.1]
plt.pie(chart,labels=labels,colors=colors,explode=explode,startangle=100,counterclock=False,autopct="%.2f%%")
plt.axis("equal")
plt.show()
```
